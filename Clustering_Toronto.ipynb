```python
wiki_toronto = requests.get('http://zims-en.kiwix.campusafrica.gos.orange.com/wikipedia_en_all_nopic/A/List_of_postal_codes_of_Canada:_M').text
wiki_toronto=BeautifulSoup(wiki_toronto,'lxml')
table_data_toronto = str(wiki_toronto.table)
display_html(table_data_toronto,raw=True)
df_toronto = pd.read_html(table_data_toronto)
df_toronto=df_toronto[0]
```

```python
df_toronto.shape
```

```python
# Dropping 'Not assigned' rows
df_toronto.drop(df_toronto[df_toronto['Borough'] == 'Not assigned'].index, inplace = True)
df_toronto.shape
```

```python
#Merger df_toronto and ll_data
df_toronto_ll = pd.merge(df_toronto,ll_data,on='Postcode')
df_toronto_ll.head()
```

```python
df_toronto_ll.shape
```

```python
toronto_geo = folium.Map(location=[43.651070,-79.347015],zoom_start=11)

for latitude,longitude,borough,neighbourhood in zip(df_toronto_ll['Latitude'],df_toronto_ll['Longitude'],df_toronto_ll['Borough'],df_toronto_ll['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [latitude,longitude],
    radius=5,
    popup=label,
    color='green',
    fill=True,
    fill_color='green',
    fill_opacity=0.6,
    parse_html=False).add_to(toronto_geo)
toronto_geo
```

```python
k=5
toronto_clustering = df_toronto_ll.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df_toronto_ll.insert(0, 'Cluster_Labels', kmeans.labels_)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for latitude, longitude, neighbourhood, cluster in zip(df_toronto_ll['Latitude'], df_toronto_ll['Longitude'], df_toronto_ll['Neighbourhood'], df_toronto_ll['Cluster_Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
